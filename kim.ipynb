```python
#Make tickers
tickers = ['AAPL', 'GOOGL', 'AMZN']
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
years = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
dummy = pd.DataFrame(columns= ["calendarYear","symbol","period", "revenue", "netIncome"])

#Create empty lists
ticker_name = []
date = []
period = []
annual = []
revenue = []


#Create Loop to run throughURL & save info into lists
for annual in tickers:
    annual_url = f"https://financialmodelingprep.com/api/v3/income-statement/{annual}?limit=120&apikey={api_key}"
    response = requests.get(annual_url).json()
    data = pd.json_normalize(response)
    data["tickers"] = annual
    data = data[["calendarYear","symbol","period", "revenue", "netIncome"]]
    #dummy = pd.concat([dummy, data1])
    #dummy.columns=["calendarYear","symbol","period", "revenue", "netIncome"]
    
print (data)    
    
```

```python
#Get Quarterly Statements
for quarter in tickers:
    quarter_url = f"https://financialmodelingprep.com/api/v3/income-statement/{quarter}?period=quarter&limit=400&apikey={api_key}"
    q_response = requests.get(quarter_url).json()
    data1 = pd.json_normalize(q_response)
    data1["tickers"] = quarter
    data1 = data1[["calendarYear","symbol","period", "revenue", "netIncome"]]
    print(data1)
```

```python
#Get Statement Balance for Cash on Hand Info & create another loop
for cash in tickers:
    cash_url = f"https://financialmodelingprep.com/api/v3/balance-sheet-statement/{cash}?apikey={api_key}&limit=120"
    c_response = requests.get(cash_url).json()
    data2 = pd.json_normalize(c_response)
    data2["tickers"] = cash
    data2 = data2[["calendarYear","symbol","period", "cashAndCashEquivalents"]]
    print(data2)
```
